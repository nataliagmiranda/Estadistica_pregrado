## **3. Medidas descriptivas**

## **3.1 Medidas de tendencia central**

Son medidas que nos ayudan a resumir una gran cantidad de información en un solo numero

## **Media:**
Es el promedio de todos los datos, puede ser susceptible a valores atípicos

**Media** = $$ \frac{1}{N} \sum_{i=1}^N x_i  $$

## **Mediana:**
Es el dato central es decir tiene la misa cantidad de datos a su izquierda y derecha, no es lo mismo que la media

### **Mediana(n=impar)** = $$ x_{(n+1)/2}^{\text{ordered}} $$

### **Mediana(n=par)** = $$ \frac{x_{(n/2)}^{\text{ordered}} + x_{(n/2)+1}^{\text{ordered}}}{2} $$

## **Moda:**
Es el dato que mas se repite, la moda no aplica para datos numéricos continuos

### **Moda** = $$ x_k $$ donde $$ \text{Freq}(x_k) = \max{(\text{Freq}(x_i))} $$

```python
#Media
media_GDP =datos['GDP'].mean()
print("Media de producto interno bruto (GDP):", media_GDP)
#Mediana
mediana_GDP =datos['GDP'].median()
print(f"Mediana de producto interno bruto (GDP): {mediana_GDP}")
#Moda
moda_GDP = datos['GDP'].mode()
print("Moda de producto interno bruto (GDP):", moda_GDP)
```

Es importante resaltar que si no hay valores que se repiten, todos los datos son considerados la moda

# **3.2 Medidas de variabilidad**

```python
#Varianza
varianza_GDP =datos['GDP'].var()
print("La varianza de producto interno bruto (GDP):", varianza_GDP)
#Desviación estandar
desvest_GDP =np.sqrt(varianza_GDP)
print("La desviación estándar de producto interno bruto (GDP):", desvest_GDP)
#Coeficiente de variación
CV_GDP=(desvest_GDP/media_GDP)*100
print("El coeficiente de variación de producto interno bruto (GDP):", CV_GDP)
#Rango
Rango_GDP=datos['GDP'].max()-datos['GDP'].min()
print("El rango de producto interno bruto (GDP):", Rango_GDP)
```

# **3.3 Medidas de forma**

```python
#Coeficiente de Asimetría
Asimetria_GDP=datos['GDP'].skew()
print("El coeficiente de asimetría de producto interno bruto (GDP):", Asimetria_GDP)
#Coeficiente de Kurtosis
Kurtosis_GDP=datos['GDP'].kurt()
print("El coeficiente de kurtosis de producto interno bruto (GDP):", Kurtosis_GDP)
```

# 4.1 Tablas y gráficas para una variable continua: para esta sección elegimos GDP


```python
#Tabla de frecuencias
# Realizando tabla de frecuencia según el numero de rangos
numero_de_rangos = 10 #el número de rangos es aproximadamente la raiz cuadrada del número de muestras

# Calcular la tabla de frecuencia absoluta con bins
Tabla_frecuencia = pd.cut(datos['GDP'], bins=numero_de_rangos, include_lowest=True, right=False).value_counts().reset_index()
Tabla_frecuencia.columns = ['Intervalo', 'Frecuencia Absoluta']

# Calcular la frecuencia acumulada
Tabla_frecuencia['Frecuencia Acumulada'] = Tabla_frecuencia['Frecuencia Absoluta'].cumsum()


# Mostrar la tabla de frecuencia
print("Tabla de Frecuencia:")
print(Tabla_frecuencia)
```

```python
# Realizando tabla de frecuencia para una variable (Almacenada en una columna)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(18, 6))
plt.bar(Tabla_frecuencia['Intervalo'].astype(str), Tabla_frecuencia['Frecuencia Absoluta'], color='skyblue')
plt.xlabel('Intervalo')
plt.ylabel('Frecuencia Absoluta')
plt.title('Tabla de Frecuencia')
plt.xticks(rotation=75)  # Rotar etiquetas del eje x si son largas
plt.show()
```
